## understanding the 'genre' column:
  the 'genres' column is in a wierd format. it is in the form of a list of dictionaries. So, by seeing it we can say that we only care about the genre name. so we must pull out the value of name key in each dictionary. we can do it manually, but the problem is... here, the list of dictionaries is in the form of a string. so, to remove this we use asl.literal_eval() method 

```python
movies['genres'][0]
```

```python
import ast
```

```python
def convert(obj):
    tags= []
    for i in ast.literal_eval(obj):
        tags.append(i['name'])
    return tags
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

## the same problem is repeated in 'keywords' column too. so repeat the same process

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

## understanding the 'cast' column:
the 'cast' is also in weird format same as 'keywords','genres' columns. so we apply the same logic for this column too. but we only consider 5 cast members for our tags.

```python
movies.cast[0]
# we just consider top 5 cast and their real names.
```

```python
def convert5(obj):
    tags= []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 5:
            tags.append(i['name'])
            count += 1
        else:
            break
    return tags
```

```python
movies['cast'] = movies['cast'].apply(convert5)
```

```python
movies.head()
```

## understanding the 'crew' column:
the column is also in a weird format. but we only consider director as the only important tag. so we find the value with job=director and return the name.

```python
def director(obj):
    tags= []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            tags.append(i['name'])
    return tags
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head()
```

## understanding the overview column:
the overview column has summary of the story of the movie(string). so, first we convert it into a List of strings.

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

### there should'nt be any spaces. As the spaces will be considered as a new tag.

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head()
```

## we create a tags column and add the tags from each column

```python
movies['tags'] =  movies['overview'] + movies['genres'] + movies['keywords']+ movies['cast'] + movies['crew']
```

```python
movies.head()
```

## we create a new DataFrame with columns{movie_id, title, tags}

```python
df = movies[['movie_id','title','tags']]
```

```python
df.head()
```

## the tags, which are of type list will be joined with spaces as a String.

```python
df['tags'] = df['tags'].apply(lambda x: ' '.join(x))
```

## all the tags are converted to lowercase to avoid confusion

## text-vectorization:
we transform the tags text into vectors.

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

## these are all the tags

```python
import sys
```

```python
np.set_printoptions(threshold=sys.maxsize)
```

```python
print(cv.get_feature_names_out())
```

## but there's a problem with them:
some words like 'action' 'acting' 'acting' are repeating. so, we must fiter them using Natural Language Processing.

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

we use PorterStemmer to do this

```python
ps = PorterStemmer()
```

```python
def stem(obj):
    li = []
    for i in obj.split():
        li.append(ps.stem(i))
    return ' '.join(li)
```

```python
stem(df['tags'][0])
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df.head()
```

```python
print(cv.get_feature_names_out())
```

## finding Similarity:
as we defined the vectors, we must find the similarity b/w them using cosine_similarity class

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
# diagonals will have 1 similarity
# similarity movie1 with movie1, similarity of movie1 with movie2,........upto 5000
```

```python
similarity.shape
```

## the recommend function:
this function takes a movie name and returns top 5 movies with highest similarity

```python
def recommend(movie):
  movie_id = df[df['title']==movie].index[0]
  distances = similarity[movie_id]
  
  movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movie_list:
    print(df.iloc[i[0]].title)
    

```

```python
recommend('Spider-Man')
```
